```python
import numpy 
titanic_train = pandas.read_csv("titanic_train.csv") 
print(titanic_train.head()) 
age = titanic_train["Age"] 
```

```python
age_isnull = pandas.isnull(age) 
print(age_isnull)  
```

```python
age_null_true = age[age_isnull]  
print(age_null_true)  
print(len(age_null_true))  
```

```python
good_ages = age[age_isnull == False]
print(good_ages)
```

```python
age_average = sum(good_ages)/len(good_ages)
print(age_average)
print(age.mean())
```

```python
row_index60_age = titanic_train.loc[60,"Age"]
print(row_index60_age)
```

```python
passenger_class = [1,2,3]
fare_by_class = {}
for this_class in passenger_class:
    pclass_rows = titanic_train[titanic_train["Pclass"] == this_class]
    pclass_fares = pclass_rows["Fare"]
    fare_by_class[this_class] = pclass_fares.mean()
print(fare_by_class)
```

```python
print(help(pandas.pivot_table))
```

```python
passenger_survival = titanic_train.pivot_table(index="Pclass", values="Fare")
print(passenger_survival)
```

```python
passenger_survival_2 = titanic_train.pivot_table(index="Embarked",values=["Fare","Survived"],aggfunc=numpy.sum) 
print(passenger_survival_2)
```

```python
def not_null_count(column):
    column_null = pandas.isnull(column)
    null = column[column_null]
    return len(null)
column_null_count = titanic_train.apply(not_null_count)
print(column_null_count)
```
